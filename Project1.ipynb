### 4.Describe statistics of fatality_count column

```python
a=df['fatality_count'].mean()
print('mean=',a)
b=df['fatality_count'].median()
print('median=',b)
c=df['fatality_count'].mode()
print("mode=",c)
d=df['fatality_count'].std()
print('standerd deviation=',d)
```

### 5.Histogram for the injury_count column

```python
df.hist (column='injury_count',grid=False,edgecolor='blue')
plt.show()
```

### 6.Explore the coorealtion between fatality_count and injury_count   

```python
correlation=(df['fatality_count'].corr(df['injury_count']))
correlation
```

### 7.Create a pie chart for the injury_count column .

```python
injury=df['injury_count'].value_counts()
plt.pie(injury.values,labels=injury.index,autopct='%1.1f%%')
plt.show()
```

### 8.Use a scatter plot to detect outliers

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x="longitude",y="gazeteer_distance",linewidths=0.5)
plt.show()
```

### 9.Plot a qq plot for the column gazeteer_distance

```python
import statsmodels.api as sm
a=df['gazeteer_distance']
sm.qqplot(a,line='s')
plt.show()
```

### 10.Create a heat map

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(),annot=True,cmap='coolwarm')
plt.show
```
